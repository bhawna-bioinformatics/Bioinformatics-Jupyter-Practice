```python
from Bio.SeqUtils import gc_fraction, molecular_weight
```

```python
from Bio.Seq import Seq
```

```python
seq=Seq("ATTCTGCTGACAATCTATCTGAAAAATTGGAAAGAGAATGGGATAGAGAG")
```

```python
print(seq)
```

```python
gc = gc_fraction(seq) * 100
print(gc)
```

```python
print(f"GC Content: {gc:.2f}%")
```

```python
mw = molecular_weight(seq, seq_type='DNA')  
print(f"Molecular Weight: {mw:.2f} Da")
```

```python
mw = molecular_weight(seq, seq_type='DNA')  # or 'RNA', 'protein'
print(mw)
```

```python
mw = molecular_weight(seq, seq_type='protein') 
print(f"Molecular Weight: {mw:.2f} Da")
```
